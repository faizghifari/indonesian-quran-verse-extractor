```python
def breakdown_labels(txt):
    labels_relevance = []
    labels_irrelevance = []
    labels_all = []
    if pd.notna(txt):
        labels = txt.split(',')
        for label in labels:
            temp_label = label.split(':')
            labels_all.append(temp_label[0])
            if (temp_label[1] == 'relevance'):
                labels_relevance.append(temp_label[0])
            else:
                labels_irrelevance.append(temp_label[0])
    else:
        labels_all.append(np.nan)
        labels_relevance.append(np.nan)
        labels_irrelevance.append(np.nan)
    return labels_all, labels_relevance, labels_irrelevance

from sklearn.metrics import classification_report, accuracy_score, hamming_loss

def print_evaluation_report(y_pred, y_test):
    print(classification_report(y_test,y_pred, digits=5))
    print("accuracy: ", accuracy_score(y_test,y_pred))
    print("hamming loss: ", hamming_loss(y_test, y_pred))
```

```python
raw_labels = pd.Series(raw_uji['labels'])

labels_relevance = []
labels_irrelevance = []
labels_all = []

for labels in raw_labels:
    temp_la, temp_lr, temp_lir = breakdown_labels(labels)
    labels_all.append(temp_la)
    labels_relevance.append(temp_lr)
    labels_irrelevance.append(temp_lir)
```

```python
raw_uji['labels_all'] = labels_all
raw_uji['labels_relevance'] = labels_relevance
raw_uji['labels_irrelevance'] = labels_irrelevance
```
